# Transferring Files

## KLC - Local workstation
These programs allow simple drag-and-drop to transfer files.
- Cyberduck
- MobaXterm

Instructions using Cyberduck as an example:
1. Select "Open Connection"
![Cyberduck Open Connection](images/cyberduck_1.png)
2. Select the SFTP protocol
3. Enter the server address (e.g. any KLC login node address: `klc0202.quest.northwestern.edu`), leave the port as 22
4.  Use your NU netid and password for login
![Cyberduck Login](images/cyberduck_2.png)



## KLC - Onedrive or external platform
`Globus` is a secure, reliable tool for transferring large files between systems — ideal for moving research data between Kellogg servers and external platforms like Northwestern Quest or cloud storage.

It supports high-speed transfers, automatic retries, and sharing data with collaborators — no need to zip files or worry about timeouts.

Detailed instructions about `Globus` are documented on NU IT Knowledge Base:
- [Globus Endpoints and Collections](https://services.northwestern.edu/TDClient/30/Portal/KB/ArticleDet?ID=1970)
- [Using the Quest Globus endpoint](https://services.northwestern.edu/TDClient/30/Portal/KB/ArticleDet?ID=1962); Collection: `Northwestern Quest`
- [Using the OneDrive/SharePoint Globus Endpoint](https://services.northwestern.edu/TDClient/30/Portal/KB/ArticleDet?ID=1969); Collection: `Northwestern OneDrive/SharePoint - High Assurance` [`link`](https://app.globus.org/file-manager?origin_id=b65388b3-daa7-48dd-ac31-b3f951ddbc93&two_pane=false)
